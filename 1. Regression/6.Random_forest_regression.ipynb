# Random Forest Regression

Random Forest is a powerful ensemble learning method that combines the predictions of multiple
decision trees to improve accuracy and robustness. It effectively reduces overfitting and enhances
the model's performance compared to a single decision tree.

# Key Concepts

*   **Ensemble Learning :**  Ensemble learning involves combining multiple models to create a more accurate and robust
prediction. Random Forest is an ensemble method that aggregates the results of multiple decision
trees

*   **Bagging (Bootstrap Aggregating) :**
Bagging involves creating multiple datasets by randomly sampling the original dataset with
replacement. Each decision tree is trained on one of these datasets. This technique helps to reduce
variance and improve model stability.

*  **Random Feature Selection :**  During the construction of each decision tree, a random subset of features is considered at each
split. This promotes diversity among the trees and reduces correlation between them.

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestRegressor
regressor =  RandomForestRegressor(n_estimators = 10 , random_state = 0)
regressor.fit(X ,y)
```

## Predicting a new result

```
regressor.predict([[6.5]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
