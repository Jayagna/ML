# R Squared

---



image_2024-08-06_112853518.png

R-squared, also known as the coefficient of determination, is a statistical measure that represents the proportion of the variance for a dependent variable that is explained by the independent variables in a regression model. It provides an indication of the goodness of fit of the model.

Residual Sum of Squares (SSres) is generally smaller in most cases than Total Sum of Squares (SStot). So, R squared is somewhere between 0 & 1.
The better our model fits the data the smaller SSres would be and the greater R Squared would be.

**Rule of Thumb (for R S quared values) :**

1.0 = Perfect fit

~0.9 = Very good

<0.7 = Not great

<0.4 = Terrible

<0 = Model makes no sense at all

# Adjusted R Squared

---



image_2024-08-06_130948404.png

Adjusted R-squared is a modified version of R-squared that has been adjusted for the number of predictors in the model. It accounts for the model complexity and provides a more accurate measure of the goodness of fit when multiple predictors are involved. Adjusted R-squared can increase or decrease depending on whether the addition of a predictor improves the model sufficiently to outweigh the penalty of adding more terms.
